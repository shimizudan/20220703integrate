## 問題

曲線 $\begin{cases}
x=\tan\theta\\
y=\cos2\theta
\end{cases} \left(0\leqq \theta\leqq\dfrac{\pi}{4}\right)$ と$x$軸，$y$軸で囲まれた部分を$x$軸を中心に1回転してできる立体の体積を求めよ。

```julia
using Plots
xₜ(θ) = tan(θ)
yₜ(θ) = cos(2θ)
plot(xₜ, yₜ, 0, π/4,fill=(0,:skyblue))
```

$$\int_0^{1}\pi y^2 dx=\pi\int_0^{\frac{\pi}4}\cos^22\theta\cdot\dfrac{d\theta}{\cos^2\theta}=\dfrac{\pi\left(4-\pi\right)}{2}$$

```julia
using MathLink,MathLinkExtras

weval(W`Integrate[π(Cos[2θ]/Cos[θ])^2,{θ,0,π/4}]`)
```
